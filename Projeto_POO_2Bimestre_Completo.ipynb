## PARTE 5 ‚Äì Projeto Final e Organiza√ß√£o

**Tarefas:**
- Consolidar todas as partes do projeto em um √∫nico script organizado.

**C√≥digo:**

```python
# Desenvolva aqui o c√≥digo da Parte 5

import mysql.connector

# ===== CONEX√ÉO COM O BANCO DE DADOS =====
conexao = mysql.connector.connect(
    host="localhost",
    user="root",
    password="",
    database="sistema_notas_leonardo"
)
cursor = conexao.cursor()


class Disciplina_Leonardo:
    def __init__(self, nome, nota, peso):
        self.nome = nome
        self.__nota = nota
        self.__peso = peso


    def get_nota_Leonardo(self):
        return self.__nota

    def set_nota_Leonardo(self, nova_nota):
        if 0 <= nova_nota <= 10:
            self.__nota = nova_nota
        else:
            print("Nota inv√°lida! Digite entre 0 e 10.")

    def get_peso_Leonardo(self):
        return self.__peso

    def set_peso_Leonardo(self, novo_peso):
        if novo_peso > 0:
            self.__peso = novo_peso
        else:
            print("Peso inv√°lido! Deve ser maior que zero.")

   
    def calcular_ponderada_Leonardo(self):
        return self.get_nota_Leonardo() * self.get_peso_Leonardo()

    def exibir_dados_Leonardo(self):
        print(f"[NORMAL] {self.nome} - Nota: {self.__nota}, Peso: {self.__peso}")

    def salvar_no_banco(self):
        cursor.execute(
            "INSERT INTO disciplinas (nome, nota, peso, tipo) VALUES (%s, %s, %s, NULL)",
            (self.nome, self.__nota, self.__peso)
        )
        conexao.commit()


class DisciplinaExtra_Leonardo(Disciplina_Leonardo):
    def __init__(self, nome, nota, peso, tipo):
        super().__init__(nome, nota, peso)
        self.tipo = tipo

    def exibir_dados_Leonardo(self): 
        print(f"[{self.tipo.upper()}] {self.nome} - Nota: {self.get_nota_Leonardo()}, Peso: {self.get_peso_Leonardo()}")

    def alterar_tipo_Leonardo(self, novo_tipo):
        self.tipo = novo_tipo

    def salvar_no_banco(self):
        cursor.execute(
            "INSERT INTO disciplinas (nome, nota, peso, tipo) VALUES (%s, %s, %s, %s)",
            (self.nome, self.get_nota_Leonardo(), self.get_peso_Leonardo(), self.tipo)
        )
        conexao.commit()



def carregar_disciplinas():
    disciplinas = []
    cursor.execute("SELECT nome, nota, peso, tipo FROM disciplinas")
    for nome, nota, peso, tipo in cursor.fetchall():
        if tipo:
            d = DisciplinaExtra_Leonardo(nome, float(nota), int(peso), tipo)
        else:
            d = Disciplina_Leonardo(nome, float(nota), int(peso))
        disciplinas.append(d)
    return disciplinas

def atualizar_nota_banco(nome, nova_nota):
    cursor.execute("UPDATE disciplinas SET nota = %s WHERE nome = %s", (nova_nota, nome))
    conexao.commit()

def menu_Leonardo():
    while True:
        print("\n===== MENU DO SISTEMA DE NOTAS - LEONARDO =====")
        print("1. Cadastrar disciplina normal")
        print("2. Cadastrar disciplina extra")
        print("3. Exibir todas as disciplinas")
        print("4. Calcular m√©dia ponderada geral")
        print("5. Alterar nota de uma disciplina")
        print("0. Sair")

        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            nome = input("Nome da disciplina: ")
            nota = float(input("Nota: "))
            peso = int(input("Peso: "))
            d = Disciplina_Leonardo(nome, nota, peso)
            d.salvar_no_banco()
            print("Disciplina normal cadastrada!")

        elif opcao == "2":
            nome = input("Nome da disciplina extra: ")
            nota = float(input("Nota: "))
            peso = int(input("Peso: "))
            tipo = input("Tipo (Eletiva, Complementar, etc): ")
            e = DisciplinaExtra_Leonardo(nome, nota, peso, tipo)
            e.salvar_no_banco()
            print("Disciplina extra cadastrada!")

        elif opcao == "3":
            disciplinas = carregar_disciplinas()
            if not disciplinas:
                print("Nenhuma disciplina cadastrada.")
            else:
                print("\n--- Disciplinas Cadastradas ---")
                for d in disciplinas:
                    d.exibir_dados_Leonardo()

        elif opcao == "4":
            disciplinas = carregar_disciplinas()
            if not disciplinas:
                print("Nenhuma disciplina para calcular m√©dia.")
            else:
                soma_ponderada = sum(d.calcular_ponderada_Leonardo() for d in disciplinas)
                soma_pesos = sum(d.get_peso_Leonardo() for d in disciplinas)
                media = soma_ponderada / soma_pesos if soma_pesos > 0 else 0
                print(f"M√©dia ponderada geral: {media:.2f}")

        elif opcao == "5":
            nome = input("Digite o nome da disciplina: ")
            nova_nota = float(input("Digite a nova nota: "))
            atualizar_nota_banco(nome, nova_nota)
            print("Nota atualizada no banco de dados!")

        elif opcao == "0":
            print("Saindo... at√© logo!")
            break

        else:
            print("Op√ß√£o inv√°lida.")


menu_Leonardo()

cursor.close()
conexao.close()





-- Criar banco de dados
CREATE DATABASE IF NOT EXISTS sistema_notas_leonardo;
USE sistema_notas_leonardo;

-- Criar tabela de disciplinas
CREATE TABLE IF NOT EXISTS disciplinas (
    id INT AUTO_INCREMENT PRIMARY KEY,
    nome VARCHAR(100) NOT NULL,
    nota DECIMAL(4,2) NOT NULL,
    peso INT NOT NULL,
    tipo VARCHAR(50) DEFAULT NULL  -- NULL para disciplinas normais
);
```
